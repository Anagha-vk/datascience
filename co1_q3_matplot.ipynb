```python
import matplotlib.pyplot as pt
x=[2,4,6,8,10]
y=[3,4,8,7,10]
pt.plot(x,y,c="orange",label="time")
pt.plot(y,x,label="speed")
pt.legend()
pt.show()
```

```python
import matplotlib.pyplot as pt
import numpy as np
x=np.linspace(0,20)
y=np.sin(x)
z=np.cos(x)
pt.plot(x,y,c="red",label="sin")
pt.plot(x,z,c="black",label="cos")
pt.legend()
pt.show()
```
